####    9.  Выберите верные утверждения:
    
* Наиболее дорогие в среднем автомобили в топ-100 - это автомобили производства Maybach. 
    
* Среди автомобилей топ-100 нет автомобилей с механической коробкой передач (MANUAL). 
    
* Все автомобили топ-100 потребляют одну категорию топлива (Engine Fuel Type). 
    
* Среди автомобилей топ-100 нет автомобилей старше 8 лет. 
    
* Большинство автомобилей топ-100 имею 12 цилиндров. 
    
* У большинства автомобилей топ-100 большие колеса. 

```python
# 1
answer = top_prices.groupby('Make')['MSRP'].mean().index.get_loc('Maybach') == 0
print("1. {0}".format(answer))

answer = 'MANUAL' not in top_prices['Transmission Type'].value_counts().index.tolist()
print("2. {0}".format(answer))

answer = top_prices['Engine Fuel Type'].nunique() == 1
print("3. {0}".format(answer))

answer = len(top_prices[(2020 - top_prices['Year']) > 8]) == 0
print("4. {0}".format(answer))

answer = top_prices['Engine Cylinders'].value_counts().index.get_loc(12.0) == 0
print("5. {0}".format(answer))

answer = top_prices['Vehicle Size'].value_counts().index.get_loc('Large') == 0
print("6. {0}".format(answer))
```

####    10. Постройте распределения расхода топлива по городу (city mpg) для автомобилей с ручной коробкой передач и автоматической. 
Не забудьте про подписи осей, размеры шрифтов, легенду.

```python
data_manual=car[car['Transmission Type'] == 'MANUAL']['city mpg']
data_automatic=car[car['Transmission Type'] == 'AUTOMATIC']['city mpg']

fig, ax = plt.subplots()
ax.hist(data_manual, bins=15, histtype='step', label='manual', density=True)
ax.hist(data_automatic, bins=15, histtype='step', label='automatic', density=True)
ax.legend(loc='upper right')
plt.xlabel('City mpg', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Distributions of fuel consumptions in city')
plt.savefig('Chetverikov_fig10.png');
plt.show()
```

####    11. Постройте зависимость средней стоимости автомобиля от числа цилиндров. 
Используйте функцию plt.errorbar, чтобы помимо средних значений отобразить среднеквадратичное отклонение величины.

```python
data=car.groupby('Engine Cylinders')['MSRP'].mean()
err=car.groupby('Engine Cylinders')['MSRP'].std()
plt.errorbar(x=data.index.tolist(), y=data, yerr=err, fmt='ro-');
plt.xlabel('Cylinders count',fontsize=14)
plt.ylabel('MSRP',fontsize=14)
plt.title('Manufacturer\'s suggested retail price\'s dependence from cylinders count');
plt.savefig('Chetverikov_fig11.png');
```

####    12. Постройте boxplot, отображающий распределение стоимости автомобилей следующих марок: Audi, Toyota, BMW, Mazda, Hyundai. 
Не забудьте про подписи осей, размеры шрифтов.

```python
make_audi=car[(car['Make'] == 'Audi')]['MSRP'].values
make_toyota=car[(car['Make'] == 'Toyota')]['MSRP'].values
make_bmw=car[(car['Make'] == 'BMW')]['MSRP'].values
make_mazda=car[(car['Make'] == 'Mazda')]['MSRP'].values
make_hyundai=car[(car['Make'] == 'Hyundai')]['MSRP'].values

plt.boxplot([make_audi, make_toyota, make_bmw, make_mazda, make_hyundai],
            labels=['Audi','Toyota', 'BMW', 'Mazda', 'Huyndai'])
plt.tick_params(axis='both', which='major', labelsize=14)
plt.ylabel('MSRP', fontsize=14);
plt.title('Boxplot of MSRP distribution for brends', fontsize=16);
plt.savefig('Chetverikov_fig12.png');
```
