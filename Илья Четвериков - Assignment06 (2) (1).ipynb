 2) $$ X \sim R[0,a] \: f(x)=\frac{1}{a} \: \forall x \in [0,a]$$

 3) $$ X \sim N(0,\sigma^2) \: f(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{x^2}{2\sigma^2}}$$

###  2. Критерии информативности (2 балла)
На лекциях говорилось, что критерий информативности для набора объ-
ектов R вычисляется на основе того, насколько хорошо их целевые переменные предсказываются константой (при оптимальном выборе этой константы):
$$
H(R)=\frac{1}{|R|}\sum_{(x_i,y_i)\in R} L(y_i,c),
$$
где $L(y, c)$ - некоторая функция потерь и
$$
c=argmin_cH(R).
$$
 
Соответственно, чтобы получить вид критерия при конкретной функции потерь, необходимо аналитически найти оптимальное
значение константы и подставить его в формулу для H(R).
Выведите критерии информативности для следующих функций потерь:

* $L(y,c)=\sum_{k=1}^K (c_k-[y=k])^2$,

* $L(y,c)=-\sum_{k=1}^K [y=k] \log c_k$. 

1) $$H(R)=\min_{\sum_{k=1}^K c_k = 1}\frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K (c_k-[y=k])^2)$$
$$ L = \frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K (c_k-[y=k])^2) + \lambda (\sum_{k=1}^K c_k - 1)$$
$$    \frac{\partial L}{\partial c_k} = \frac{2\sum_{(x_i,y_i)\in R}(c_k - [y=k])}{|R|} + \lambda = 0 $$
$$    2(\frac{\sum_{(x_i,y_i)\in R}c_k}{|R|}-p_k)+\lambda=0 $$
$$    2(c_k-p_k)+\lambda=0 $$
Суммируем по $k$:
$$ 2(1-1)+K\lambda=0$$
$$ \lambda=0$$
$$ c_k = p_k - достигается минимум$$
Подставляем $p_k$ в $H(R)$:
$$ H(R) = \frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K (p_k-[y=k])^2) = \frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K (p_k^2-2[y=k]p_k+[y=k]^2) = \frac{1}{|R|} \sum_{k=1}^K \sum_{(x_i,y_i)\in R}(p_k^2-2[y=k]p_k+[y=k]) = 
\frac{1}{|R|} \sum_{k=1}^K (|R|p_k^2-2p_k|R|p_k+|R|p_k) = \sum_{k=1}^K (p_k^2-2p_k^2+p_k) = \sum_{k=1}^K p_k (1-p_k)$$

2) $$H(R)=\min_{\sum_{k=1}^K c_k = 1}-\frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K [y=k] \log c_k$$
$$ L = -\frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K [y=k] \log c_k + \lambda (\sum_{k=1}^K c_k - 1) $$
$$    \frac{\partial L}{\partial c_k} = -\frac{\sum_{(x_i,y_i)\in R}[y=k]}{|R|c_k} + \lambda = 0 $$
$$ -\frac{p_k}{c_k}+\lambda = 0$$
$$ c_k=\frac{p_k}{\lambda}$$
Суммируем по $k$:
$$ 1=\frac{1}{\lambda}$$
$$ \lambda = 1 \; c_k = p_k $$
Подставим $p_k$ в $H(R)$:
$$ H(R)=-\frac{1}{|R|}\sum_{(x_i,y_i)\in R} \sum_{k=1}^K [y=k] \log p_k=-\frac{1}{|R|}\sum_{k=1}^K  \log p_k\sum_{(x_i,y_i)\in R}[y=k]= -\sum_{k=1}^K p_k \log p_k$$

### 3.  Переобучение и неустойчивость решающих деревьев (3 балла)
* сгенерируйте случайную выборку $X=\{x_1,x_2\}$ с количеством объектов $N=100$: $x_1,x_2$ из нормального распределения, целевая переменная $y=\sqrt{x_1^2+x_2^2}$;

* изобразите выборку на плоскости, по осям - значения признаков $x_1$, $x_2$, цвет отражает значение целевой переменной;

* постройте решающее дерево для задачи регрессии с параметрами по умолчанию;

* с помощью get_grid постройте разделяющую границу регрессора;

* постройте деревья-регрессоры  для разных значений глубины дерева и минимального числа объектов в листе;

* изобразите на одном графике  соответствующие разделяющие границы регрессоров (например, с помощью plt.subplot2grid);

* Решающие деревья — это алгоритмы, неустойчивые к изменениям обучающей выборки, т.е. при малейших её изменениях итоговый классификатор может радикально измениться. Посмотрите, как будет меняться структура дерева при обучении на разных 90%-х подвыборках (постройте три регрессора и соответствующие разделяющие границы регрессоров)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeRegressor

def get_grid(data): 
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), # создаем координатную матрицу c шагом 0.01
                         np.arange(y_min, y_max, 0.01))

N = 100
X = np.random.randn(N,2)
y = np.sqrt(X[:,0]**2 + X[:,1]**2)

plt.scatter(X[:,0],X[:,1], c=y)
plt.xlabel("x1")
plt.ylabel("x2");
plt.title("Samples");

```

```python
regr = DecisionTreeRegressor()
regr.fit(X, y)
xx, yy = get_grid(X)
predicted = regr.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 8))
plt.pcolormesh(xx, yy, predicted, cmap='ocean')
plt.scatter(X[:,0],X[:,1], c=y)
plt.xlabel("x1")
plt.ylabel("x2");
plt.title("TreeRegressor Bound");
```

```python

max_depth_array =[2, 4, None]
min_samples_leaf_array =[1,2,15]

fig = plt.figure(figsize=(15, 5))
for max_depth, min_samples_leaf, i in zip(max_depth_array, min_samples_leaf_array, [0,1,2]):
    regr = DecisionTreeRegressor(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
    regr.fit(X, y)
    xx, yy = get_grid(X)
    predicted = regr.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax = plt.subplot2grid((1, 3), (0, i))
    ax.pcolormesh(xx, yy, predicted, cmap='ocean')
    ax.scatter(X[:,0],X[:,1], c=y)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title('max_depth={} and min_samples_leaf={}'.format(max_depth, min_samples_leaf))
    
fig.suptitle("Tree Regressor with different parameteres", fontsize=16);
```

```python

fig = plt.figure(figsize=(15, 5))
for i in [0,1,2]:
    regr = DecisionTreeRegressor()
    rows = np.random.choice(X.shape[0], size=90)
    X_sub = X[rows, :]
    y_sub = y[rows]
    regr.fit(X_sub, y_sub)
    xx, yy = get_grid(X)
    predicted = regr.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax = plt.subplot2grid((1, 3), (0, i))
    ax.pcolormesh(xx, yy, predicted, cmap='ocean')
    ax.scatter(X[:,0],X[:,1], c=y)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title('subsample {}'.format(i))
    
fig.suptitle("Tree Regressor with different subsamples", fontsize=16);
```

### 4. Решающее дерево для регрессии на данных (2 балла)
* разделите выборку на обучающую и тестовую в соотношении $7:3$;
* с помощью GridSearch подберите наилучшие параметры для решающего дерева регрессора (выведите параметры наилучшего регрессора). 

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
data = load_boston()

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split

scoring='r2'

max_features=[None, 'log2', 'sqrt']

max_depth =[2, 4, 6, 8, 10, 20, 50]

min_samples_leaf=np.arange(1,20)

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.30, random_state=42)
param_grid = {'max_features':max_features, 'max_depth':max_depth, 'min_samples_leaf':min_samples_leaf}
gs = GridSearchCV(DecisionTreeRegressor(), param_grid, scoring=scoring, cv=10)
gs.fit(X_train, y_train)
print("Best params: {}".format(gs.best_params_))
y_pred = gs.predict(X_test)

print ('R2 score =', metrics.r2_score(y_test, y_pred))
```
